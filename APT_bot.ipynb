```python
import requests
import lxml
from bs4 import BeautifulSoup
import re
from itertools import groupby
import pandas as pd
```

```python
source = requests.get('https://www.olx.pl/nieruchomosci/mieszkania/sprzedaz/krakow/?search%5Bfilter_float_price_per_m%3Ato%5D=7000&search%5Bfilter_enum_market%5D%5B0%5D=secondary&search%5Bfilter_float_m%3Afrom%5D=60').text
soup = BeautifulSoup(source, 'lxml')

def print_val():
    lst = []
    for div in soup.find_all('div', class_ = 'offer-wrapper'):
        
        for loc in div.find_all('span'):
            loc = loc.text.rstrip('\n \n')
            lst.append(loc)
        for a in div.find_all('a', href=True): 
            if a.text: 
                lst.append(a['href'])
        for price in div.find_all('p', class_ = 'price'):
            price = price.text
            x = re.findall('[0-9]+', price)
            x = int(''.join(x))
            lst.append(x)
            

    return lst
            
def clear_list(lst):
     
    for i in lst:
        if i == '' or i == ' ' or i == '#':
            lst.remove(i)
    lst = list(dict.fromkeys(lst))
    lst = [list(v) for k,v in groupby(lst, key = lambda x: type(x) != int) if type(k) != int]
    lst_loc = lst[::2]
    
    lst_price = lst[1::2]
    return lst_loc, lst_price

def create_df(a,b):
    df = pd.DataFrame(list(zip(a, b),), columns = ['lokalizacja', 'cena'])
    return df

def main():
    lst1 = print_val()
    lst_a, lst_b = clear_list(lst1)
    return create_df(lst_a, lst_b)
        
```

```python
t = main()
```
